# Declaração de variáveis 

```python
ano_atual = 2019
ano_fabricacao = 2003
km_total = 44410.0

km_media = km_total / (ano_atual - ano_fabricacao)
km_media
```

```python
#Em tese, se no ano posterior o dono do veículo mantesse o mesmo padrão de uso, esteima-se que o veículo atinga a km abaixo
km_total += km_media
km_total
```

#Conclusão
"valor = valor + 1" equivale a "valor += 1'
