# Indexação "np.array"


```python
#Criando array com indice de 0, a 9 [1 a 10]
contador = np.arange(10)
contador
```

```python
contador[-1]
```

```python
item = 6
index = item - 1
contador[index]
```

```python
#unindo 2 arrays

km = np.array([44410.0, 5712.0, 37123.0, 0, 25757.0])
anos = np.array([2003, 1991, 1990, 2019, 2006])

dados = np.array([km, anos])
dados
```

```python
#[linha, indice]
dados[1, 2]
```

```python
#[:selecionando todas as linhas, selecionando do indice de 1 a 3]
dados[:, 1:4]
```

```python
#[:selecionando todas as linhas, selecionando do indice de 1 a 2] e [filtrando as linhas]
dados[:, 1:3][0] / (2019 - dados[:, 1:3][1])
```
