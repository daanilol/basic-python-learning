# Estudando Listas

```python
dados = [['rodas_de_liga', 'travas_eletricas', 'piloto_automatico', 'bancos_de_couro', 'ar_condicionado', 'sensor_de_estacionamento', 'sensor_crepuscular', 'sensor_de_chuva'], ['central_muiltimidia', 'teto_panoramico', 'freio_abs', '4_x_4', 'painel_digital', 'piloto_automatico', 'bancos_de_couro', 'camera_de_estacionamento'], ['piloto_automatico', 'controle_de_estabilidade', 'sensor_crepuscular', 'freio_abs', 'cambio_automatico', 'bancos_de_couro', 'central_multimedia', 'vidros_eletricos']]
```

```python
dados
```

# Manipulando lista 1

```python
for lista in dados:
  for item in lista:
    print(item)
```

```python
acessorios = []

for lista in dados:
  for item in lista:
    acessorios.append(item)
```

# Manipulando lista 2

```python
acessorios
```

# set() - Tirando duplicatas

```python
set(acessorios)
```

# List comprehension

```python
list_comprehension = [item for lista in dados for item in lista]
```

```python
list(set(list_comprehension))
```
