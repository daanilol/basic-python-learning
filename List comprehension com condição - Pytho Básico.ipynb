# List comprehenson com If

```python
# 1º item da lista - Nome do veículo
# 2º item da lista - Ano de fabricação
# 3º item da lista - Veículo é zero km?

dados = [
         ['jetta_variant', 2003, False],
         ['passat', 1991, False],
         ['crossfox', 1990, False],
         ['ds5', 2019, True],
         ['aston_martin_db4', 2006, False],
         ['palio_weekend', 2012, False],
         ['a5', 2019, True],
         ['serie_3_cabrio', 2009, False],
         ['dodge_jorney', 2019, False],
         ['carens', 2011, False]
]

dados
```

# Condicionando uma lista

```python
zero_km = []
for lista in dados:
  if lista[2] == True:
    zero_km.append(lista)

zero_km
```

```python
zero_km_N = []

for lista in dados:
  if lista[2] == False:
    zero_km_N.append(lista)

zero_km_N
```

# São 0 km

```python
[lista for lista in dados if lista[2] == False]
```

# Não são 0 km

```python
[lista for lista in dados if lista[2] == False]
```
