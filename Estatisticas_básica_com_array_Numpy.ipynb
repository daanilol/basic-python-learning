# Estatisticas básica com array Numpy

```python
import numpy as np
```

```python
#criando arrays
anos = np.loadtxt(fname='carros-anos.txt', dtype=int)
km = np.loadtxt(fname='carros-km.txt')
valor = np.loadtxt(fname='carros-valor.txt')
```

```python
#array unidimenssinal
anos.shape
```

```python
#transforma arrays unidimenssionais em colunas de um array bidimensional
dataset = np.column_stack((anos, km, valor))
```

```python
dataset.shape
```

```python
#analisando a media por colunas
np.mean(dataset, axis=0)
```

```python
#verificando media de colunas específicas, aqui no caso, a coluna index 1 (km)
np.mean(dataset[:, 1])
```

```python
#aqui no caso, a coluna indice 2 (valor)
np.mean(dataset[:, 2])
```

ndarray.sum()

```python
#soma por colunas
dataset.sum(axis=0)
```

```python
#soma todas as linhas de uma coluna especifica
dataset[:, 1].sum()
```

np.sum()

```python
np.sum(dataset[:, 1])
```

```python
np.sum(dataset[:, 2])
```
